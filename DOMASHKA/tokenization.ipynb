```python
text_raw = open("Jeeves and Wooster stories", "r").readlines()
text = ""
for t in text_raw:
    text += t+"\n"

text
```

BlanklineTokenizer -  разбивает текст, рассматривая в качестве разделителя любую последовательность пустых строк (содержащих только символы пробела или табуляции)

```python
start_time = timeit.default_timer()

word_tok = nltk.tokenize.BlanklineTokenizer()
print(word_tok.tokenize(text))

print(timeit.default_timer() - start_time)
```

LegalitySyllableTokenizer - разбивает слова по слогам, основываясь на принципе правильности и максимизации наборов.

```python
with open('Jeeves and Wooster stories', 'r', encoding='utf-8') as f:
    all_line = f.readlines()
clear_words = [line.split('\t')[0].lower() for line in all_line if len(line.split('\t')) > 1]
print(len(clear_words))
```
