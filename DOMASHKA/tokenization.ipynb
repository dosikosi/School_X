BlanklineTokenizer - разбивает текст, рассматривая в качестве разделителя любую последовательность пустых строк (содержащих только символы пробела или табуляции)

```python
start_time = timeit.default_timer()

word_tok = nltk.tokenize.BlanklineTokenizer()
print(word_tok.tokenize(text))

print(timeit.default_timer() - start_time)
```

LegalitySyllableTokenizer - Разбивает слова по слогам, основываясь на принципе правильности и максимизации наборов.Принцип правильности (Legality Principle) – это не зависящий от используемого языка принцип, утверждающий, что начала и окончания слогов (не включая гласный звук) допустимы только в том случае, если они встречаются в качестве начал или окончаний слов в языке.

```python
with open("Jeeves and Wooster stories.txt") as file:
    text = file.read()
tokenizer = LegalitySyllableTokenizer(tokenized_source_text = text)
tokens = tokenizer.tokenize(text)

print(tokens)
```

LineTokenizer - для разделения строки использует символ новой строки (при необходимости может отбрасывать пустые строки). Его применение похоже на применение метода s.split(‘\n’).

```python
start_time = timeit.default_timer()

word_tok = nltk.tokenize.LineTokenizer()
print(word_tok.tokenize(text))

print(timeit.default_timer() - start_time)
```

MWETokenizer - берет строку, которая уже была разделена на токены, и повторно токенизирует ее, объединяя выражения из нескольких слов в отдельные токены, используя словарь MWE. Это может быть полезно в том случае, когда в тексте встречаются устойчивые выражения или именованные сущности, состоящие из нескольких слов.

```python
with open("Jeeves and Wooster stories.txt") as file:
    text = file.read()
tokenizer = MWETokenizer()
tokens = tokenizer.tokenize(text)

print(tokens)
```

NLTKWordTokenizer - является «деструктивным» токенизатором. То есть, применяемые регулярные выражения приводят исходный текст в состояние, не поддающееся восстановлению

```python
with open("Jeeves and Wooster stories.txt") as file:
    text = file.read()
tokenizer = NLTKWordTokenizer()
tokens = tokenizer.tokenize(text)

print(tokens)
```

PunktSentenceTokenizer - разделяет текст на список предложений, используя алгоритм обучения без учителя (Unsupervised learning) для построения модели слов-сокращений, словосочетаний и слов, с которых начинаются предложения.

```python
start_time = timeit.default_timer()

word_tok = nltk.tokenize.PunktSentenceTokenizer()
print(word_tok.tokenize(text))

print(timeit.default_timer() - start_time)
```

RegexpTokenizer - разбивает исходный текст на подстроки, используя регулярное выражение, передаваемое ему в качестве параметра.

```python
from nltk.tokenize import RegexpTokenizer

with open("Jeeves and Wooster stories.txt") as file:
    text = file.read()

tokenizer = RegexpTokenizer(r'\w+')
tokens = tokenizer.tokenize(text)

print(tokens)
```

ReppTokenizer - Токенизатор с использованием алгоритма REPP, описанного в статье « Tokenization: Returning to a Long Solved Problem – A Survey, Contrastive Experiment, Recommendations, and Toolkit» (авторы Rebecca Dridan и Stephan Oepen (2012)). Стоит отметить, что ссылка на статью в оригинальной документации неактуальна.

```python
import nltk

with open("Jeeves and Wooster stories.txt") as file:
    text = file.read()

tokens = nltk.word_tokenize(text)

print(tokens)
```

SExprTokenizer - используется для поиска в исходном тексте выражений в скобках. В частности, он делит строку на последовательность подстрок, являющихся выражениями в скобках (включая любые вложенные выражения в скобках) или другими токенами, разделенными пробелами.

```python
with open("Jeeves and Wooster stories.txt") as file:
    text = file.read()
tokenizer = SExprTokenizer()
tokens = tokenizer.tokenize(text)

print(tokens)
```

SpaceTokenizer - Разбивает строку на токены, используя в качестве разделителя пробел. Результат работы этого метода полностью совпадает с результатом метода s.split(‘ ‘).

```python
with open("Jeeves and Wooster stories.txt") as file:
    tokens = SpaceTokenizer().tokenize(file.read())

print(tokens)
```

StanfordSegmenter (устарел и выдает ошибку, некоторые языки требуют более обширной предварительной обработки токенов, которая обычно называется сегментацией. Например, StanfordSegmenter предназначен для «токенизации» или «сегментации» слов китайского или арабского текста (также можно использовать для английского, французского и испанского языков).)

```python
start_time = timeit.default_timer()

word_tok = nltk.tokenize.StanfordSegmenter()
print(word_tok.tokenize(text))

print(timeit.default_timer() - start_time)
```

SyllableTokenizer - Это еще один токенизатор для разбиения слов на слоги. Его работа основана на принципе последовательности звучания (Sonority Sequencing Principle, SSP). Алгоритм SSP так же, как и алгоритм Legality Principle не зависит от используемого языка. 

```python
with open("Jeeves and Wooster stories.txt") as file:
    tokens = SyllableTokenizer().tokenize(file.read())

print(tokens)
```

TabTokenizer - Разбивает строку на токены, используя в качестве разделителя пробел, только для разделения строки использует символ табуляции, что совпадает с применением метода s.split(‘\t’)

```python
with open("Jeeves and Wooster stories.txt") as file:
    text = file.read()
tokenizer = TabTokenizer()
tokens = tokenizer.tokenize(text)

print(tokens)
```

TextTilingTokenizer (в тексте нет явных разрывов параграфов) - метод автоматического разделения полноразмерных текстовых документов на последовательные блоки, состоящие из нескольких абзацев, которые представляют собой отрывки или подтемы. Алгоритм предполагает, что в ходе описания подтемы используется определенный набор слов, и когда подтема изменяется, значительная часть словарного запаса также изменяется.

```python
from nltk.tokenize import TextTilingTokenizer

with open("Jeeves and Wooster stories.txt", encoding="utf-8") as file:
    tokens = TextTilingTokenizer(w=8, k=10, stopwords='ONE').tokenize(file.read())

print(tokens)
```

ToktokTokenizer -  это простой универсальный токенизатор, на вход которого подается по одному предложению в строке. Согласно документации он был протестирован и дает достаточно хорошие результаты для английского, персидского, русского, чешского, французского, немецкого, вьетнамского, таджикского и некоторых других языков. Входные данные должны быть представлены в кодировке UTF-8.

```python
with open("Jeeves and Wooster stories.txt") as file:
    text = file.read()
tokenizer = ToktokTokenizer()
tokens = tokenizer.tokenize(text)

print(tokens)
```

TreebankWordTokenizer - использует регулярные выражения для разделения текста на слова так, как это реализовано в Penn Treebank. Для этого последовательно выполняются следующие шаги:
разделяются стандартные сокращения, например, «don’t» -> »do n’t» или «they’ll» -> «they ‘ll» для английского языка;
большинство знаков препинания, разделяются на отдельные токены;
разделяются запятые и одинарные кавычки, если за ними следует пробел;
отделяются точки, которые проставлены в конце строк.

```python
import timeit
start_time = timeit.default_timer()

word_tok = nltk.tokenize.TreebankWordTokenizer()
print(word_tok.tokenize(text))

print(timeit.default_timer() - start_time)
```

TweetTokenizer - Это токенизатор с поддержкой особенностей коротких сообщений Twitter, разработанный для гибкой и простой адаптации к новым предметным областям и задачам. Основная логика его работы заключается в следующем:
Определяется список регулярных выражений, для поиска в исходном тексте гиперссылок, смайликов, телефонных номеров, имен пользователей Twitter, хештегов и др.
Регулярные выражения помещаются по порядку в скомпилированный объект регулярного выражения, называемый word_re.
Выполняется токенизация с помощью метода word_re.findall(s), где s – строка, исходного текста.

```python
with open("Jeeves and Wooster stories.txt") as file:
    tokens = TweetTokenizer().tokenize(file.read())

print(tokens)
```

WhitespaceTokenizer - При помощи этого токенизатора исходный текст разделяется на группы символов, ограниченных пробелами (обычно, это отдельные слова, либо слова со следующими за ними знаками препинания)

```python
start_time = timeit.default_timer()

whitespace = nltk.tokenize.WhitespaceTokenizer()
print(whitespace.tokenize(text))

print(timeit.default_timer() - start_time)
```

WordPunctTokenizer - Преобразует текст в последовательность буквенных и неалфавитных символов с помощью регулярного выражения ‘\w+|[^\w\s]+’. Отличается от WhitespaceTokenizer тем, что отделяет знаки препинания и иные символы пунктуации от слова, за / перед которым они идут, и возвращает их в качестве отдельных токенов.

```python
with open("Jeeves and Wooster stories.txt") as file:
    tokens = WordPunctTokenizer().tokenize(file.read())

print(tokens)
```

лексическое многообразие 

```python
def lexical_mnogoobrazie(text):
    tokens = text.split()
    return len(set(tokens)) / len(tokens)

with open("Jeeves and Wooster stories.txt", encoding="utf-8") as file:
    text = file.read()

mnogoobrazie = lexical_mnogoobrazie(text)
print(f"Лексическое многообразие: {mnogoobrazie:.2f}")
```

собственный токенизатор

```python
class Tokenizer:
    def __init__(self):
        pass

    def tokenize(self, text):
        text = text.lower()
        text = text.replace(',', ' ')
        tokens = text.split()
        return tokens

tokenizer = Tokenizer()

with open("Jeeves and Wooster stories.txt") as file:
    text = file.read()
    tokens = tokenizer.tokenize(text)

print(tokens)

```
