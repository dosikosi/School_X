```python
class Tokenizer:
    def __init__(self):
        pass

    def tokenize(self, text):
        text = text.lower()
        text = text.replace(',', ' ')
        tokens = text.split()
        return tokens

tokenizer = Tokenizer()

with open("Jeeves and Wooster stories.txt") as file:
    text = file.read()
    tokens = tokenizer.tokenize(text)

print(tokens)

```
